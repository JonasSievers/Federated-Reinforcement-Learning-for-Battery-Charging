# Merge Weather Dataset and Household datasets (PV, Grossload, TotalLoad)

```python
# Required libraries
import pandas as pd
```

```python
# Read the weather data.
weather = pd.read_csv('../2010-2013 Weather_preprocessed.csv', )

try:
    weather.drop(columns=["Unnamed: 0"], inplace=True)
except:
    print("No Unnamed column")

display(weather.head(2))
display(weather.tail(2))
```

```python
def merge_dataframes(df, weather):
    # Merge the weather and load dataframes
    merged_df = pd.merge(df, weather, left_on='Date', right_on='time')

    # Remove the redundant "time" column
    merged_df.drop('time', axis=1, inplace=True)
    return merged_df
```

# Merged_Weather_PV_data

```python
pv_df = pd.read_csv('../2010-2013 PV_preprocessed.csv', )
```

```python
# Check the shape
print('Weather data shape: ',weather.shape)
print('Load data shape: ',pv_df.shape)

merged_weather_pv_df = merge_dataframes(pv_df, weather)

display(merged_weather_pv_df.shape)
display(merged_weather_pv_df.head(2))
display(merged_weather_pv_df.tail(2))
```

```python
# Save final dataset
merged_weather_pv_df.to_csv('../2010-2013 Merged_Weather_PV_dataset.csv', index=False)
```
