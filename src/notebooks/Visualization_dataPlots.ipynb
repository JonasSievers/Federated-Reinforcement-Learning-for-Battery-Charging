```python
energy_data = pd.read_csv("../../data/3final_data/Final_Energy_dataset.csv").fillna(0).set_index('Date')
energy_data.head(1)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Assuming energy_data is your DataFrame and it's already loaded
# Convert index to datetime if not already
if not isinstance(energy_data.index, pd.DatetimeIndex):
    energy_data.index = pd.to_datetime(energy_data.index)

# Define the figure size
plt.figure(figsize=(9, 3))

start = 48*11
end = start + 48*7

# Plot load and pv for user 1 and 2 with predefined colors
plt.plot(energy_data.index[start:end], energy_data["load_11"][start:end]*2, label="Load Building 11", color='#294586')
plt.plot(energy_data.index[start:end], energy_data["pv_11"][start:end]*2, label="PV Building 11", color='#D7C000')

# Adding title and labels (optional)
plt.xlabel("Time", fontsize=16)
plt.ylabel("Power [kW]", fontsize=16)

# Formatting the date labels
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d \n %H:%M'))

# Ensure exactly 7 x-ticks are shown
# Get the current x-axis limits
xlims = ax.get_xlim()
# Generate 7 equally spaced ticks within these limits
tick_positions = pd.date_range(start=energy_data.index[start], end=energy_data.index[end], periods=5)
ax.set_xticks(tick_positions)

# Format and rotate date labels
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Displaying the legend
plt.legend(fontsize=16, loc="upper left")

# Ensure the layout is tight so labels do not overlap
plt.tight_layout()

# Save the plot as a PDF
plt.savefig("fig/load_pv_patterns_building_11.pdf", format='pdf')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

energy_data.index = pd.to_datetime(energy_data.index)

# Define the figure and axes
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 3), sharex=True)

# Define start and end indices
start = 48*11
end = start + 48

# Plot price on the first subplot
ax1.plot(energy_data.index[start:end], energy_data["price"][start:end], label="Price", color='#294586', zorder=-100)
ax1.set_xlabel("Time", fontsize=16, color='black')
ax1.set_ylabel("Price [$]", fontsize=16, color='black')
ax1.tick_params(axis='y', labelcolor='black', labelsize=14)
ax1.set_title("Price", fontsize=18)

# Plot emissions on the second subplot
ax2.plot(energy_data.index[start:end], energy_data["emissions"][start:end], label="Emissions", color='#D7C000', zorder=-100)
ax2.set_xlabel("Time", fontsize=16, color='black')
ax2.set_ylabel("Emissions [kg]", fontsize=16, color='black')
ax2.tick_params(axis='y', labelcolor='black', labelsize=14)
ax2.set_title("Emissions", fontsize=18)

# Formatting the date labels
for ax in [ax1, ax2]:
    # Set major locator and formatter
    ax.xaxis.set_major_locator(mdates.AutoDateLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d \n %H:%M'))
    
    # Get the current x-axis limits
    xlims = ax.get_xlim()
    
    # Generate a list of dates for the ticks
    # Ensure that there are exactly 3 tick positions
    ticks = pd.date_range(start=energy_data.index[start], end=energy_data.index[end], periods=3)
    ax.set_xticks(ticks)
    
    # Set tick labels
    ax.set_xticklabels([tick.strftime('%Y-%m-%d \n %H:%M') for tick in ticks], fontsize=14)

# Adding legends to each subplot
ax1.legend(fontsize=16, loc='upper left')
ax2.legend(fontsize=16, loc='upper left')

# Ensure the layout is tight so labels do not overlap
fig.tight_layout()

# Save the plot as a PDF
plt.savefig("fig/emission_price_patterns.pdf", format='pdf')

# Display the plot
plt.show()
```
